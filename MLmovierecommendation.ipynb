MOVIE RECOMMENDATION SYSTEM

Importing the libraries

```python
import numpy as np
import pandas as pd
```

Uploading the dataset

```python
df=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Number of rowns and columns in the dataframe



```python
df.shape
```

```python
df.info()
```

Selecting important features

```python
selected_features=['Genre','Description','Actors','Director']
```

Checking for null values

```python
df.isnull().sum()
```

```python
df.info()
```

Replacing the null values with null string

```python

for feature in selected_features:
  df[feature] = df[feature].fillna('') 
```

```python
revenue_mean=np.mean(df['Revenue (Millions)'],axis=None)
df['Revenue (Millions)']=df['Revenue (Millions)'].fillna(revenue_mean)
df['Revenue (Millions)']=df['Revenue (Millions)'].astype(int)
```

```python
revenue_mean=np.mean(df['Metascore'],axis=None)
df['Metascore']=df['Metascore'].fillna(revenue_mean)
df['Metascore']=df['Metascore'].astype(int)
```

```python
df.isnull().sum()
```

Combining the features

```python
combined_features = df['Genre']+' '+df['Description']+' '+df['Director']+' '+df['Actors']
```

```python
print(combined_features)
```

Converting the text data to feature vectors

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Getting a movie name from the user 

```python
movie_name = input(' Enter your favourite movie name : ')
```

Creating a list withh all the movie names given in the dataset

```python
titles = df['Title'].tolist()
```

Finding a close match for the movie entered by the user 

```python
find_match = difflib.get_close_matches(movie_name, titles)
```

```python
close_match =find_match[0]
```

Finding the index of the movie with title

```python
index_of_the_movie = df[df.Title ==close_match]['Rank'].values[0]
```

Getting the index of the movie along with similar codes 

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))

len(similarity_score)
```

Sorting the movies based on the similarity score





```python
sorted_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_movies)
len(sorted_movies)
```

Printing the recomended movies 

```python
print('Movies suggested for you : \n')

i = 999
for movie in sorted_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['Title'].values[0]
  if (i>990):
    print(10-i%90, '.',title_from_index)
    i-=1
```
